# Common DEGs between Chronic Heart Failure and Covid Patients

```python
import numpy as np
import pandas as pd
```

```python
df_chf_down = pd.read_excel('chronicHeartFailureDEG.xlsx',sheet_name='downRegulated')
```

```python
df_chf_down.shape
```

```python
df_chf_up = pd.read_excel('chronicHeartFailureDEG.xlsx',sheet_name='upRegulated')
```

```python
df_chf_up.head(2)
```

```python
df_chf_up.shape
```

```python
df_covid_down = pd.read_excel('covidDEGs.xlsx',sheet_name='downRegulated')
```

```python
df_covid_down.head(2)
```

```python
df_covid_down.shape
```

```python
df_covid_up = pd.read_excel('covidDEGs.xlsx',sheet_name='upRegulated')
```

```python
df_covid_up.head(2)
```

```python
df_covid_up.shape
```

```python
df_chf_covid_down = pd.merge(df_chf_down,df_covid_down,how='inner',left_on='GENE_SYMBOL',right_on='GENE_SYMBOL')
```

```python
df_chf_covid_up = pd.merge(df_chf_up,df_covid_up,how='inner',left_on='GENE_SYMBOL',right_on='GENE_SYMBOL')
```

```python
df_chf_covid_down
```

```python
df_chf_covid_down[['GENE_SYMBOL','GENE_SYMBOL']]
```

```python
df_chf_covid_up
```

```python
df_chf_covid_up[['GENE_SYMBOL','GENE_SYMBOL']]
```

```python
df_output_down = df_chf_covid_down['GENE_SYMBOL']
```

```python
df_output_up = df_chf_covid_up['GENE_SYMBOL']
```

```python
df_output_all = pd.merge(df_chf_covid_down,df_chf_covid_up,how='outer')
```

```python
df_output_all.head(2)
```

```python
df_all = df_output_all[['P.Value','logFC','GENE_SYMBOL']]
```

```python
writer = pd.ExcelWriter('covidCHF_commonDEGs.xlsx',engine='xlsxwriter')
```

```python
writerPvalLogfc = pd.ExcelWriter('CHF_covid_all.xlsx',engine='xlsxwriter')
```

```python
df_output_down.to_excel(writer,sheet_name='downRegulated')
```

```python
df_output_up.to_excel(writer,sheet_name='upRegulated')
```

```python
df_all.to_excel(writerPvalLogfc)
```

```python
writer.save()
writer.close()
writerPvalLogfc.save()
writerPvalLogfc.close()
```
